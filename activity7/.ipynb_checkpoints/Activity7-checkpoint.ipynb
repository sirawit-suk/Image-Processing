```python
import numpy as np
import cv2
from keras.preprocessing.image import  ImageDataGenerator
from keras.preprocessing.image import img_to_array
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('Grid_image.jpg')
src2 = cv2.imread('kaimuk.jpg')
img2 = cv2.cvtColor(src2,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(src)
```

# 7.1 Effects of Interpolation 

```python
def interpolate(src):
    reduce_factors =  np.array([2,8,15])
    scale_factors = 1/reduce_factors
    inter_methods = [cv2.INTER_NEAREST, cv2.INTER_LINEAR,cv2.INTER_CUBIC,cv2.INTER_AREA]
    meth_label = ["INTER_NEAREST","INTER_LINEAR","INTER_CUBIC","INTER_AREA"]
    
    
    f,ax = plt.subplots(3,4,figsize=(20,20))

    y = 0
    for sf in scale_factors:
        x = 0
        for meth in inter_methods:
            im = cv2.resize(src,((int)(src.shape[0]*sf),(int)(src.shape[1]*sf)),interpolation=meth)
            ax[y,x].set_title(meth_label[meth])
            ax[y,x].imshow(im)
            x += 1
        y += 1
    f.show()
```

```python
interpolate(src)
```

```python
interpolate(img2)
```

# 7.2 Image Augmentation

```python
def write_video(file_path, frames, fps):
    w, h = frames[0].shape[1], frames[0].shape[0]   # 300, 300
    fourcc = cv2.VideoWriter_fourcc('m', 'p', '4', 'v')
    writer = cv2.VideoWriter(file_path, fourcc, fps, (w, h))

    for frame in frames:
        writer.write(frame)

    writer.release()
# ref: https://www.programcreek.com/python/example/72134/cv2.VideoWriter
```

```python
use_image = cv2.cvtColor(img2,cv2.COLOR_RGB2BGR)
use_image = img_to_array(use_image)
use_image = np.expand_dims(use_image, 0)
fill_method = ['constatn','nearest','reflect','wrap']
frames = []
```

```python
use_image.shape
```

```python
Npic = 9
rotation_range = 90
width_shift_range = [-200,200]
height_shift_range = 0.0
shear_range = 0.0
zoom_range  = 0.0
horizontal_flip = False
```

```python
for m in fill_method:
    datagen = ImageDataGenerator(
                                    featurewise_center=False, samplewise_center=False,
                                    featurewise_std_normalization=False, samplewise_std_normalization=False,
                                    zca_whitening=False, zca_epsilon=1e-06, rotation_range=0, width_shift_range=0.0,
                                    height_shift_range=0.0, brightness_range=None, shear_range=0.0, zoom_range=0.0,
                                    channel_shift_range=0.0, fill_mode=m, cval=0.0,
                                    horizontal_flip=False, vertical_flip=False, rescale=None,
                                    preprocessing_function=None, data_format=None, validation_split=0.0, dtype=None
                                )
    pic = datagen.flow(use_image,batch_size = 1)
    for i in range(1,Npic):
        batch = pic.next()
        im_result = batch[0].astype('uint8')
        frames.append(im_result)
```

```python
write_video('test.mp4', frames, 3)
```
