```python
import cv2
import numpy as np
import math
import matplotlib.image as mpimg

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

```python
path = './logo.jpg'
src = cv2.imread(path)
src2 = plt.imread(path)
```

## 1.1 Color Reordering

```python
src_gray_scale = src # Already BGR
B = src_gray_scale[:, :, 0]
G = src_gray_scale[:, :, 1]
R = src_gray_scale[:, :, 2]

f, axarr = plt.subplots(1, 4, figsize=(20,20)) 
axarr[0].title.set_text('Original')
axarr[0].imshow(src)

axarr[1].title.set_text('Blue')
axarr[1].imshow(B, cmap='gray')

axarr[2].title.set_text('Green')
axarr[2].imshow(G, cmap='gray')

axarr[3].title.set_text('Red')
axarr[3].imshow(R, cmap='gray')
```

```python
src_gray_scale = src2
R = src_gray_scale[:, :, 0]
G = src_gray_scale[:, :, 1]
B = src_gray_scale[:, :, 2]

f, axarr = plt.subplots(1, 4, figsize=(20,20)) 
axarr[0].title.set_text('Original')
axarr[0].imshow(src2)

axarr[1].title.set_text('Red')
axarr[1].imshow(R, cmap='gray')

axarr[2].title.set_text('Green')
axarr[2].imshow(G, cmap='gray')

axarr[3].title.set_text('Blue')
axarr[3].imshow(B, cmap='gray')
```
